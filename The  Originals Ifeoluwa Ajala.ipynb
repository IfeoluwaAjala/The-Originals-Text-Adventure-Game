```python
"""
   DocString:
   
   A) Introduction:
   This game is based on the series the Originals (2013 - 2018) 
   https://en.wikipedia.org/wiki/The_Originals_(TV_series). The series 
   involves the first vampire family and the vampires they sired. They had 
   friends who were witches and werewolf. It consists of three stages, and 
   has defined functions for start, win, and lose.The game is about survival 
   and if you have watched the series already good luck!
   
   Round 1: Pick a character.
   Round 2: Successfully answer 3 questions from Marcel.
   Round 3: Defeat the original vampires!
   
   B) Known Bugs and/or Errors:
   Random questions are sometimes repeated, remove function entirely removes
   the question.
   
"""

#Importing necessary packages exit and random for the game
from sys import exit
import random

#defining the function start which begins the game.
def start():
    print("""
          You awake, open your eyes and find yourself in a thick forest. 
          The only thing you remember is that you need to find your cousin 
          at the annual ball in New Orleans.
          """)
    input("<Press enter to continue>\n") #this is to transition to starting phase.
    starting_phase() #this is defined below

#defining the function starting_phase which is stage 1 and printing a question with options.
def starting_phase():
    print("""
    What character would you like to be?
    1) Human
    2) Vampire
    3) Werewolf
    """)        
 
    choice = input(">   \n").lower() #prompting for choice input
    
#coding an if loop
    if "1" in choice or "human" in choice:
        print("\nYou're off to New Orleans!\n") #print statement
        input("<Press enter to continue>\n") #prompting enter for transition
        city_gate() #this moves us to stage 2 of the game
        
    elif "2" in choice or "vampire" in choice:
        print("""
        Well now I guess the game is over, since we don't kill our kind!
        """)
        input("<Press enter to continue>\n") #prompting enter for transition
        starting_phase() #this moves us back to stage 1 of the game
        
    elif "3" in choice or "werewolf" in choice:
        print("\nNow that you are a werewolf, you need a disguise ;)")
        input("<Press enter to continue>\n") #prompting enter for transition
        city_gate() #this moves us to stage 2 of the game
        
    else:
        print("Invalid command. Please try again.") #controlling for potential errors.
        input("<Press enter to continue>\n") #prompting enter for transition
        starting_phase() #this moves us back to stage 1 of the game

#defining the function city_gate which is stage 2 and printing a statement
def city_gate():
    print(""" 
    As you are about to enter the city gates to French quarters,
    a vampire Marcel sired by Klaus jumps in front of you and shouts:
    
    "Wait! Before you enter the city, you need to answer these three 
    questions, or I will have you for dinner."
                     
    Try your best to answer these three questions truthfully.
    """)
    
    #declaring a question list
    question_list = ['What is your name?',
                    'Are you human?',
                    'New Orleans is the capital of?',
                    'Whose guest are you?']
    questions = 3
    
    
#coding the while loop function with conditional statements
    while questions > 0: 
        question = random.choice(question_list) #using the random choice function to select random choices from the question_list
        input("<Press enter to continue>\n") #prompting enter for transition
        print(question) 
        print(" ")
        
        if question == question_list[0]:
            print("You can input anything and you're ok.") #printing a statement
            print(" ")
            name = input("Your name:  \n").lower() #prompting a name input with lower function to convert input
            print(f"Marcel: Nice to meet you {name.capitalize()}!") #dynamically applying user input
            questions -=1
            
        elif question == question_list[1]:
            print("1) I am human.")
            print("2) I am a vampire.")
            print("3) I am a wolf.")
            quest = input(">  \n").capitalize() #controlling for potential errors

            #start of a nested conditional
            if quest == '1':
                print("Thanks for letting me know. I am glad to make a new friend.\n")
                questions -=1 #to reduce the questions asked after each random question
                      
            elif quest == '2':
                print("\nWell now I guess the game is over, since we don't kill our kind!")
                input('<Game Over>')
                starting_phase() #this takes us back to stage 1 of th game
            
            elif quest == '3':
                print("\nNice to meet you, I suppose you are from Harley's pack")
                      
            else:
                print("Invalid command. Please try again.") #controlling for errors
                input("<Press enter to continue>\n") 
    
            #end of nested conditional
                      
        elif question == question_list[2]:
            print("1) Louisiana ")
            print("2) Arkansas  ")
            
            capital = input(">  \n").lower() #prompting an input 
            
            #start of a nested conditional
            if '1' in capital or 'louisiana' in capital:
                print("Marcel: Ok, ok, that is correct!. \n") #print statement
                questions -= 1 #reducing the choices to be made 
                
            elif '2' in capital or 'arkansas' in capital:
                print("Marcel: That is incorrect, I get to eat you!\n")
                input('<Game Over>')
                starting_phase() #this takes us back to stage 1 of the game
            
              #end of nested conditional
            
        elif question == question_list[3]:
            print("1) Attending without an invite")
            print("2) Elijah")
            print("3) Klaus\n")
                      
            host = input(">  \n").lower() #prompting an input
            
            #start of a nested conditional
            if '1' in host or 'attending' in host or 'without' in host:
                print("I get to eat you!")
                fail() #this function defined below signals a failure in the game
            
            elif '2' in host or 'Elijah' in host:
                print("You got lucky, I like Elijah")
                print("Marcel disappears and you see your cousin Jack walk towards you")
                winning() #function defined below which indicates winning
                      
            elif '3' in host or 'Klaus' in host:
                print("Marcel: Huh? ... Klaus Mikaelson.")
                print("Marcel disappears because he is afraid of Klaus, you are free to attend the annual ball.\n")
             #end of a nested conditional which flows to the final stage of the game         
                break
                      
    print("Congratulations! You made it to the annual ball!\n")
    input("<Press enter to continue>\n")
    annual_ball() #this function signals the third stage of the game.

    #defines the function annual_ball and prints statement
def annual_ball():
    print("""
You are at the annual ball and Mikaelson brothers challenge you to a fight for
your life.
    """)
    print("What do you do?")
    print("1) Attack with a cross")
    print("2) Run away!")
    print("3) Spray them with gasoline and light em up")
    
    survival = input(">    \n").lower() #prompts input and lower() controls for potential errors
    
    #coding an if loop
    
    if '1' in survival or 'attack' in survival or 'cross' in survival:
        print("You attack with a cross and kill Elijah.") #print statement
        print("Klaus yells and runs away to fight another day.\n")
        input("<Press enter to continue>\n")
        winning() #this indicates success in the game
                
    elif '2' in survival or 'run' in survival:
        print('Klaus catches you in an instant and bites you.') #print statement
        print('You have lost the game!') #print statement
        fail() #this function defined below signals a failure in the game
    
    elif '3' in survival or 'spray' in survival:
        print("While you doused Elijah with gasoline, Klaus took his brother and escaped!")
        print("You end up back at the city gate") #print statement
        input("<Press enter to continue>\n")
        city_gate() #this fucntion takes us back to stage 2 of the game

#defining the winning function used above
def winning():
    print(f"""{'=-=-=' * 19}
You have completed your escape from the oldest vampires at New Orleans! Great job!
{'=-=-=' * 19} 
          """)
    print("""
    
 /$$     /$$                                                      /$$                 /$$   /$$    /$$ /$$
|  $$   /$$/                                                     | $$                |__/  | $$   | $$| $$
 \  $$ /$$//$$$$$$  /$$   /$$       /$$$$$$/$$$$   /$$$$$$   /$$$$$$$  /$$$$$$        /$$ /$$$$$$ | $$| $$
  \  $$$$//$$__  $$| $$  | $$      | $$_  $$_  $$ |____  $$ /$$__  $$ /$$__  $$      | $$|_  $$_/ | $$| $$
   \  $$/| $$  \ $$| $$  | $$      | $$ \ $$ \ $$  /$$$$$$$| $$  | $$| $$$$$$$$      | $$  | $$   |__/|__/
    | $$ | $$  | $$| $$  | $$      | $$ | $$ | $$ /$$__  $$| $$  | $$| $$_____/      | $$  | $$ /$$       
    | $$ |  $$$$$$/|  $$$$$$/      | $$ | $$ | $$|  $$$$$$$|  $$$$$$$|  $$$$$$$      | $$  |  $$$$//$$ /$$
    |__/  \______/  \______/       |__/ |__/ |__/ \_______/ \_______/ \_______/      |__/   \___/ |__/|__/
                                                                                                          
                                                                                                          
                                                    
    """)
    input("<Press enter to exit, you champion!>\n")
    
    exit(0) #exit function to force an end to the game
    
#defining a fail function to enable a player loose in the game

def fail():
    print('\nYou have failed in your escape from the oldest vampires at New Orleans.')
    input('<Game Over>') #input statement
    exit(0)
                      
start() #calling the first function that will start the game
                      
```
